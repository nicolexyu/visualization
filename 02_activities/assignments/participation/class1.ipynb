```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import seaborn
```

```python
# set seed
np.random.seed(613)

# generate data
x = np.arange(50)
y = np.random.randint(0,100,50)

# scatter plot 
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y)
```

```python
# bar plot
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
# line plot
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
# Create the histogram
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
# Create the line plot 
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

# Set labels and title
ax.set_title("Total growth over time")
ax.set_xlabel("Total Growth")
ax.set_ylabel("Years since start")
```

```python
# Setting fonts
font1 = {"family": "sans-serif","color":"blue", "size":20}
font2 = {"family": "monospace","color":"green", "size":14}
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title("Total growth over time", fontdict = font1)
ax.set_ylabel("Total growth", fontdict = font2)
ax.set_xlabel("Years since start", fontdict = font2)
```

```python
# Activity 
# copy the provided code and attempt to replicate the output in Python

# libraries and data
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
df=pd.DataFrame({'x_values': range(1,101), 'y_values': np.random.randn(100)*15+range(1,101), 'z_values': (np.random.randn(100)*15+range(1,101))*2 })
 
# Cut your window in 1 row and 2 columns, and start a plot in the first part
plt.subplot(121)
plt.plot( 'x_values', 'y_values', data=df, marker='o', alpha=0.4)
plt.title("A subplot with 2 lines")
 
# And now add something in the second part:
plt.subplot(122)
plt.plot( 'x_values','z_values', data=df, linestyle='none', marker='o', color="orange", alpha=0.3)
 
# Show the graph
plt.show()

# Recall the aesthetic, substantive, and perceptual qualities of data visualizations. Does your visualization of choice succeed in each area?

# Aesthetic: 
# The use of color differentiation (blue and orange) in the plots makes it easy to distinguish between the two datasets.
# The markers (marker='o') make the data points visually distinct.

# Substantive:
# The visualization appears to show two different data distributions (y_values and z_values) plotted against x_values.

# Perceptua:
# The perceptual clarity is strong due to the use of simple scatterplots with distinct markers and colors.
```

```python
# Adjusting dot type
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y,
          marker ="*", # "o"/ "+"
          color = "indigo")
```

```python
# Line style
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
          marker ="*", # "o"/ "+"
          color = "indigo",
           linestyle = "--",
           linewidth = 2)
```

```python
# Marker color
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
          marker ="*", 
          color = "indigo", # line color
          linestyle = "--",
          linewidth = 2,
           markeredgecolor = "#FA9359", # marker edge color
           markerfacecolor = "#000000") # marker inside colo
```

```python
# Grid line
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
          marker ="*", 
          color = "indigo", 
          linestyle = "--",
          linewidth = 2,
           markeredgecolor = "#FA9359", 
           markerfacecolor = "#000000") 
ax.grid() # both vertical and horizontal grid lines
#ax.grid(axis = "x") # vertical grid lines
#ax.grid(axis = "y") # horizontal grid lines
```
