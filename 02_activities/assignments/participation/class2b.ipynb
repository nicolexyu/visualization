```python
# loading libraries and generating some new sample data
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
fig,(ax1, ax2) = plt.subplots(ncols=2,
nrows=1,
figsize=(7, 3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(9, 3))

font1 = {'family': 'serif', 'size': 10}
font2 = {'family': 'serif', 'size': 14}

ax1.scatter(x1, y1, marker='o', color='red')
ax2.bar(x2, y2, color='darkgreen')

# Customize ax1
ax1.set_title('Important Scatterplot', fontdict=font2)
ax1.set_ylabel('Measure', fontdict=font1)
ax1.set_xlabel('Time', fontdict=font1)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
    ['ax2', 'ax3']],
    figsize=(7, 4))
for label, ax in someaxes.items():ax.text(0.5, 0.5, f'This is {label!r}',
    fontsize=14,
    ha = 'center',
    transform=ax.transAxes)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
['ax2', 'ax3']],
figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
plt.show()
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],['ax2', 'ax3']],figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel("A Big Label", fontsize=18)
someaxes["ax2"].set_xlabel("Another Label", fontsize=18)
someaxes["ax3"].set_xlabel("Label 2: 2 Fast 2 Furious", fontsize=18)
plt.show()
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],['ax2', 'ax3']],figsize=(7, 4),layout = "constrained")
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel("A Big Label", fontsize=18)
someaxes["ax2"].set_xlabel("Another Label", fontsize=18)
someaxes["ax3"].set_xlabel("Label 2: 2 Fast 2 Furious", fontsize=18)
```
