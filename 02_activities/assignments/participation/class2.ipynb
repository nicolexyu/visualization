```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
# make a line plot with both of variables on the same axes
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
ax.plot(x,y2)
```

```python
# Add a ‘labelʼ argument to each plot object
# Call ‘legendʼ and specify our legend location
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right')
```

```python
# By adding arguments to our legend object, we can change different elements of our legend
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right',
frameon = True, #add frame around the legend
fontsize = 12, #change font size
ncol = 2, #specify number of columns
shadow = True )
```

```python
# Specifying legend location as before
# Specifying where we want to anchor that location on our figure using bbox_to_anchor
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc = "upper left",
    bbox_to_anchor =(1, 1))
```

```python
# scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# use the text() function to specify coordinates and text
ax.text(10, 95, "This value is important!")
```

```python
# modify the text annotation by adding arguments

# scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# use the text() function to specify coordinates and text
ax.text(10, 95, "This value is important!",
    ha='center', ##alignment of text
    color = 'red', ##modify font colour
    size = 20) ##modify font size
```

```python
# making a simple, empty set of axes
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

# position the annotation with reference to the data
ax.text(1, 5, ". Data:(1, 5)", transform=ax.transData )
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes )
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure )
```

```python
# specify the location to point at and the location where we want the text
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
arrowprops=dict(facecolor='black', shrink=0.05))
```

```python
# modifying arrows
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = 'lower right')

ax.annotate("This is important!",
    xy=(10, 95), xytext=(20, 94),
    arrowprops = dict(arrowstyle = "wedge",
    color = "hotpink"))
```

```python
# remove tick marks or labels from axis
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
# define a maximum number of ticks on axis
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
# set tick marks at a particular interval (5) using MultipleLocator
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
# rotate axis labels and horizontal alignment
# modifying fontsize and font
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = "lower right")
plt.xticks(rotation=45, ha="right", fontsize = 12, font = "serif")
```

```python
# modifying axis title fonts
font1 = {"family":"serif","color":"indigo"}

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = "lower right")
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.xticks(rotation=45, ha="right", fontsize = 12, font = "serif")

plt.xlabel("Shiny New X Axis!", fontsize = 18, fontdict = font1)
```
